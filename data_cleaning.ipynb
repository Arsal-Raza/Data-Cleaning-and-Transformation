# Question:1
### Remove newline characters from all columns(i.e Hits: '\n372')

```python
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

df.head()
```

# Question:2
### Some columns have 'star' characters. Strip those columns of these stars and make the columns numerical(i.e IR)

```python
# Identify columns with star characters and convert to numerical values
for col in df.columns:
    if df[col].dtype == 'object' and df[col].str.contains('★').any():
        df[col] = pd.to_numeric(df[col].str.replace('★', ''), errors='coerce')

df[['IR','W/F']]
```

# Question:3
### Do the height and weight column has appropriate data types?

##### Hint:
#####     Convert the height and weight columns to numerical forms for better analysis practice

```python
# Convert height to centimeter
def convert_height_to_cm(height):
    feet, inches = map(int, height.replace('"', '').split("'"))       #as height is like this(5'7") in dataset...
    total_inches = (feet * 12) + inches
    return total_inches * 2.54  # 1 inch is approximately 2.54 cm

df['Height'] = df['Height'].apply(convert_height_to_cm)

# Convert weight to kilogran
def convert_weight_to_kg(weight):
    pounds = int(weight.split('lbs')[0])
    return pounds * 0.453592  # 1 pound is approximately 0.453592 kg

df['Weight'] = df['Weight'].apply(convert_weight_to_kg)
```

```python
df[['Height','Weight']]        #it will return height and weight columns in centimeters and kilograms respectively
```

# Question:4
### Based on the 'Joined' column, check which players have been playing at a club for more than 10 years!

##### Hint:
#####      Separate joined column into year,month and day column and then perform analysis

```python
# Convert the 'Joined' column to a datetime object
df['Joined'] = pd.to_datetime(df['Joined'])

# Create new columns for year, month, and day
df['Joined_Year'] = df['Joined'].dt.year
df['Joined_Month'] = df['Joined'].dt.month
df['Joined_Day'] = df['Joined'].dt.day

new_df = df.set_index('LongName')[['Joined_Year','Joined_Month','Joined_Day']]
new_df
```

#### Analyze which players have been playing at a club for more than 10 years:

```python
new_df['Years_At_Club'] = new_df['Joined_Year'].max() - new_df['Joined_Year']

# Filter players who have been at the club for more than 10 years
players_more_than_10_years = new_df[new_df['Years_At_Club'] > 10]

players_more_than_10_years['Years_At_Club']
```

# Question:5
### 'Value', 'Wage' and 'Release Clause' are string columns. Convert them to numbers

```python
def convert_currency_to_number(value):
    if pd.isnull(value):
        return value
    if 'M' in value:
        return float(value.replace('€', '').replace('M', '')) * 1e6
    elif 'K' in value:
        return float(value.replace('€', '').replace('K', '')) * 1e3
    else:
        return float(value.replace('€', ''))

columns_to_convert = ['Value', 'Wage', 'Release Clause']

# Apply the conversion function to the specified columns
for col in columns_to_convert:
    df[col] = df[col].apply(convert_currency_to_number)

df[['Value','Wage','Release Clause']]
```

# Question:6
### Which players are highly valuable but still underpaid (on low wages)? 
##### hint: scatter plot between wage and value

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Filter players with high value and low wage
high_value_low_wage = df[(df['Value'] > 0) & (df['Wage'] > 0) & (df['Wage'] < df['Value'])]

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(high_value_low_wage['Wage'], high_value_low_wage['Value'], color='blue', alpha=0.7)
plt.title("High Value but Low Wage Players")
plt.xlabel("Wage")
plt.ylabel("Value")
plt.grid(True)

plt.show()
```
